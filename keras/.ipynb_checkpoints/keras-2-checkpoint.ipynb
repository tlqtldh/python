```python
import scipy
import numpy as np
import matplotlib
import pandas
import sklearn
import pydotplus
import h5py
import tensorflow
import keras
```

## 1. 문제 정의 하기
* 이진 분류 문제에 적용
* 8개의 변수를 통해 당뇨병 발병 유무가 기록된 '피마족 인더언 당뇨병 발병 데이터셋' 활용
* 8개의 변수를 독립변수로 보고 당뇨병 발병 유무를 예측하는 이진 분류 문제로 정의

## 2. 데이터 준비하기
* https://archive.ics.uci.edu/ml/datasets/Piam+Indians+Diabetes
* Dataset 항목
    * 인스턴스 수 : 768개
    * 속성 수 : 8개 (임신횟수, 포도당 농도, 이완기 혈압, 삼두근 피부 두께, 혈청인슐린, 체질량, 가족력, 나이
    * 클래스 수 : 2개 (1 발병, 0 발병하지 않음)

```python
from keras.models import Sequential
from keras.models import Sequential
from keras.layers import Dense, Activation

np.random.seed(5) # 매번 실행시 마다 결과가 달라지지 않도록 Random seed를 명시적으로 지정

dataset = np.loadtxt("./data/pima-indians-diabetes.data", delimiter=",") # '," 구분자로 dataset에 입력(CSV 형식)

x_train = dataset[:700, 0:8] # 768개의 데이터중 700개 까지 Training data로 사용
y_train = dataset[:700, 8] # 맨 마지막 filed 는 답
x_test = dataset[700:, 0:8] # 701 ~ 768 까지 Test를 위한 data로 사용
y_test = dataset[700:, 8]
```

## 3. 모델 구성하기
* 속성이 8개 이기 때문에 입력 뉴런은 8개이고, 발병 유무에 따른 이진 분류이기 때문에 출력 뉴런은 1개
* 첫번째 Dense Layer는 hidden layer로 부터 8개의 뉴런을 입력 받아 12개를 출력 Activation 함수는 relu
* 두번째 Dense Layer는 12개의 뉴런을 입력 받아 8개를 출력, Activation 함수는 relu
* 세번째 Dense Layer는 8개의 뉴런을 입력 받아 1개를 출력, Activation 함수는 Sigmoid

```python
model = Sequential()
model.add(Dense(units=12, input_dim=8, activation='relu'))
model.add(Dense(units=8, activation='relu'))
model.add(Dense(units=4, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

## 4. 모델 시각화
* Keras의 시작화 기능을 이용하여 구성된 Layer의 Vector image 형태로 볼수 있음

```python
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot

%matplotlib inline

SVG(model_to_dot(model, show_shapes=True).create(prog='dot', format='svg'))
```

## 5. 모델 학습 과정 설정
* 모델을 정의 하였다면 손실 함수와 최적화 암고리즘으로 엮어 보자
    * loss : 현재 가중치 세트를 평가하는데 사용한 손실 함수 이진 클래스의 문제이므로 'binary_crossentropy'로 지정
    * optimizer : 최적의 가중치를 검색하는데 사용하는 최적화 알고리즘으로 경사하강법의 하나인 'adam'을 지정
    * metric : 평가 척도를 나타내며 분류 문제에서는 일반적으로 'accuracy'로 지정

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

## 6. 모델 학습 시키기
* 모델을 학습 시키기 위해 fit() 함수를 사용
    * 1st parameter : 입력 변수 
    * 2nd parameter : 출력 변수 (label)
    * epochs: 전체 Training Dataset에 대해 학습 반복 힛수 
    * batch_size: 가충치를 업데이트할 배치 크기

```python
model.fit(x_train, y_train, epochs=1500, batch_size=64)
```

## 7. 모델 평가하기
* Test data를 통해 모델의 정확도를 측정

```python
scores = model.evaluate(x_test, y_test)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))
```
